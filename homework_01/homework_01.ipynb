## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
number = input('Введите число: ')
#number = '12345'
revers_num = number[0] + number[len(number)-2:0:-1] + number[len(number)-1]
print(revers_num)
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days = int(input('Введите количество дней:'))
weeks = days // 7
weekends = weeks * 2

if days % 7:
    if (days - weeks * 7) == 6:
        weekends += 1
print(weekends)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
length, width, piece = int(input()), int(input()), int(input())
#length = 4
#width = 5
#piece = 12
result = False

idx_length = 0
idx_width = 0
for idx_length in range(length):
    if (width * idx_length) == piece:
        result = True
        break
if not result:
    for idx_width in range(width):
        if (length * idx_width) == piece:
            result = True
            break
print(result)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
number = int(input())
#number = 506
number_rome = "I"*number

number_rome = number_rome.replace("I"*5, "V")
number_rome = number_rome.replace("V"*2, "X")
number_rome = number_rome.replace("X"*5, "L")
number_rome = number_rome.replace("L"*2, "C")
number_rome = number_rome.replace("C"*5, "D")
number_rome = number_rome.replace("D"*2, "M")

number_rome = number_rome.replace("DCCCC", "CM")
number_rome = number_rome.replace("CCCC", "CD")
number_rome = number_rome.replace("LXXXX", "XC")
number_rome = number_rome.replace("XXXX", "XL")
number_rome = number_rome.replace("VIIII", "IX")
number_rome = number_rome.replace("IIII", "IV")

print(number_rome)
```
